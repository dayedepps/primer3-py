```python
import os
import random
import sys
import unittest

import pandas

LOCAL_DIR = os.getcwd()
PACKAGE_DIR = os.path.dirname(LOCAL_DIR)

sys.path = [PACKAGE_DIR] + sys.path

import primer3

# Insure that we imported the local copy of primer3-py
assert primer3.__file__ == os.path.join(PACKAGE_DIR, 'primer3', '__init__.pyc')
```

```python
data = []
for _ in range(10000):
    oligo_a = ''.join([random.choice('ATGC') for _ in range(8)])
    oligo_b = ''.join([random.choice('ATGC') for _ in range(8)])
    ab = primer3.calcHeterodimer(oligo_a, oligo_b)
    ba = primer3.calcHeterodimer(oligo_b, oligo_a)
    data.append([oligo_a, oligo_b, ab.ds, ba.ds, ab.dh, ba.dh, ab.dg, ba.dg, ab.dg != ba.dg])
```

```python
df = pandas.DataFrame(data=data, columns=['oligo_a', 'oligo_b', 'ds_ab', 'ds_ba', 'dh_ab', 'dh_ba', 'dg_ab', 'dg_ba', 'dg_mismatch'])
```
